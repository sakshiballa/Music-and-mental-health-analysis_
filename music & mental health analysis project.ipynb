## Removing outliers

```python
data_index = df[df['BPM'] > 200].index
df = df.drop(data_index, axis=0)
```

```python
df.dropna(subset="BPM")
```

```python
df.drop(['Timestamp', 'Permissions'], axis=1, inplace=True)
```

## Checking correlation

```python
corr=df.corr()["Age"]
corr.sort_values()
```

```python
sns.heatmap(df.corr())
```

```python
ages=df.groupby("Age").Age.count()
ages
```

```python
plt.figure()
ages.plot(kind="bar",figsize=(12,6))
```

#### It is observed that ages 17 to 19 participants are more while ages 68 to 89 participants are the least

```python
#fav genres

genres=df.groupby("Fav genre").size()
genres
plt.figure()
genres.plot(kind="bar")
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(df["Age"],df["Fav genre"],width=0.5,align="center")
ax.set_xticks(np.arange(df["Age"].min(),df["Age"].max()+1,5))
ax.set_xlabel("Age")
ax.set_ylabel("Fav genre")
plt.show()
```

```python
plt.figure()
sns.lineplot(x=df["Fav genre"],y=df["Age"],ci=None)
plt.xticks(rotation=90)
```

### Ages 50 and above tend to listen to Gospel genre while the younger age groups of 20-25 listen to Rock the most, followed by Jazz, R&B,Classical,Metal. Ages 15 to 30 have Lofi as favourite genre

```python
df.groupby("Primary streaming service")["Primary streaming service"].count().plot.pie()
```

```python
stream_serv=df.groupby("Primary streaming service")["Primary streaming service"].count()
fig,ax=plt.subplots()
ax.pie(stream_serv,labels=stream_serv.index,autopct="%1.1f%%")
ax.set_title("primary streaming service")
plt.show()
```

### Around 62 % users use Spotify as the primary streaming service followed by youtube and apple music at 12.9% and 7%. 

```python
df.groupby(["Fav genre","Primary streaming service"]).size()
```

```python
#least popular genres
genre_count=df["Fav genre"].value_counts()
low_genres=genre_count.loc[genre_count<=25]
low_genres.keys()
```

### Country, Rap, Jazz,Lofi,,Gospel and Latin are the least played genres

```python

low_genres_df=df.loc[df["Fav genre"].isin(low_genres.keys())]
low_genres_df.groupby("Primary streaming service")["Primary streaming service"].count().sort_values(ascending=False)
```

### Even the least played genres have Spotify as the top streaming service. However, the genres don't affect the streaming platform

## Relation with mental disorders

```python
depression=df.groupby(["Depression"]).Depression.count()
ocd=df.groupby(["OCD"]).OCD.count()
anxiety=df.groupby(["Anxiety"]).Anxiety.count()
insomnia=df.groupby(["Insomnia"]).Insomnia.count()
```

```python
fig,ax =plt.subplots(2,2, figsize=(12,8))
depression.plot(kind="bar",label="Depression",ax=ax[0,0])
anxiety.plot(kind="bar",label="Anxiety",ax=ax[0,1])
ocd.plot(kind="bar",label="OCD",ax=ax[1,0])
insomnia.plot(kind="bar",label="Insomnia",ax=ax[1,1])

plt.show()
```

```python
#listening habits while working:
plt.figure(figsize=(12,8))
sns.histplot(x=df["Hours per day"],hue=df["Music effects"])
```

```python
plt.figure()
sns.boxplot(x=df["Fav genre"],y=df["Hours per day"])
plt.xticks(rotation=90)
```

### Latin and Jazz genres are listened to the most hours by the users, Gospel is listened for least hours

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["Anxiety"],errwidth=0)
plt.xticks(rotation="90")
```

### People with Anxiety tend to listen to Folk and Hip Hop more

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["Depression"],errwidth=0)
plt.xticks(rotation="90")
```

### People with Depression tend to listen to Lofi more

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["Insomnia"],errwidth=0)
plt.xticks(rotation="90")
```

### People with Insomnia listen to Lofi and Gospel the most

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["OCD"],errwidth=0)
plt.xticks(rotation="90")
```

### It is again observed that Lofi is the popular genre with OCD affected

```python
plt.figure()
sns.barplot(x=df['Fav genre'], y=df['Age'], hue=df['Music effects'], errwidth=0, palette='Reds')
plt.xticks(rotation=90)
```

### Gospel listeners' condition has improved due to listening habits , while rock has mixed effects, Rap listeners too have improved in their condition followed by Jazz

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["OCD"],hue=df["Music effects"],errwidth=0,palette="Blues")
plt.xticks(rotation=90)
```

### Classical music has worsened the OCD while Latin music has shown no effect. We can conclude that Lofi music has been a little helpful in improving Ocd.

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["Anxiety"],hue=df["Music effects"],errwidth=0,palette="Blues")
plt.xticks(rotation=90)
```

### Video game music has the worst effects on the Anxiety while Latin music has improved the condition.

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["Depression"],hue=df["Music effects"],errwidth=0,palette="Blues")
plt.xticks(rotation=90)
```

### Classical and video game music has worsened the Depression while Lofi has improved it.

```python
plt.figure()
sns.barplot(x=df["Fav genre"],y=df["Insomnia"],hue=df["Music effects"],errwidth=0,palette="Blues")
plt.xticks(rotation=90)
```

### Video game music has the worst effects on insomnia while Lofi and Gospel has improved the condition. Latin music continues to show no effect.

## Conclusion:
### The genres have wide effects on the mental health of the listeners. The correlation is impeccable. From the given observations, Video Game music and Classical music has worsened the mental health conditions overall while Lofi has been the best in the improvement of conditons. Latin music has mostly no effect with the conditions, rest other genres have mixed effects on the users, depending on the individual's condition.
